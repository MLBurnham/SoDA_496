```python
import pandas as pd
import sqlite3
```

## SQL Database Creation

```python
pokemon = pd.read_csv("https://raw.githubusercontent.com/MLBurnham/SoDA_496/main/Class%20Materials/Jupyter/pokemon.csv")

pokemon.head()
```

```python
pokemon['Type 1'].unique()
```

```python
con = sqlite3.connect('pokemon.db')

pokemon.to_sql('all_gen', con = con, index = False, if_exists = 'replace')

con.close()
```

## SQL Database Query

SELECT *columnname* FROM *tablename* WHERE *condition*

```python
con = sqlite3.connect('pokemon.db')

c = con.cursor()

c.execute("""
SELECT name
FROM sqlite_master
WHERE type='table'
""")

tables = c.fetchall()

con.close()

tables
```

```python
con = sqlite3.connect('pokemon.db')

c = con.cursor()

c.execute("""
SELECT *
FROM all_gen
WHERE ("Type 1" LIKE 'Water' OR "Type 1" LIKE 'Fire')
AND "Attack" > 100
""")

results = c.fetchall()

con.close()

results
```
